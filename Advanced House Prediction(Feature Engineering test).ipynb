# Calculating Percentage of null values

```python
features_nan=[feature for feature in data.columns if data[feature].isnull().sum()>=1]
for feature in features_nan:
    print(feature,": {} % missing values".format(np.round(data[feature].isnull().mean(),4)))
```

# Handling Missing values

```python
#lets handle categorical features

features_nan=[feature for feature in data.columns if data[feature].dtypes=='O' and data[feature].isnull().sum()>=1]

for feature in features_nan:
    print(feature,":  {} % missing".format(np.round(data[feature].isnull().mean(),4)))
```

```python
# replace missing cat var with a new label (missing)
def cat_missing(data,features_nan):
    rdata=data.copy()
    rdata[features_nan]=rdata[features_nan].fillna("Missing")
    return rdata

data=cat_missing(data,features_nan)
```

```python
data[features_nan].head(10)
```

```python
data[features_nan].isnull().sum()
```

```python
#lets find numerical variable with missing features
num_features_nan=[feature for feature in data.columns if data[feature].dtypes!='O' and data[feature].isnull().sum()>=1]

for feature in num_features_nan:
    print(feature," {} % missing".format(np.round(data[feature].isnull().mean(),4)))
```

```python

for feature in num_features_nan:
    
    median_value=data[feature].median()
    data[feature+'nan']=np.where(data[feature].isnull(),1,0)
    data[feature].fillna(median_value,inplace=True)


```

# Modifying Temporal Variables

```python
year_features=[feature for feature in data.columns if  'Year' in feature or 'Yr' in feature]
year_features
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature]=data['YrSold']-data[feature]
```

```python
data[year_features].head()
```

# Transforming Numerical Variables

we need to apply log transformation on some variables because they are skewed

```python
skewed_num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
for feature in skewed_num_features:
    data[feature]=np.log(data[feature])
    
```

```python
data
```

# Handling Rare Categrical features

```python
cat_features=[feature for feature in data.columns if data[feature].dtype=='O']
```

```python
for feature in cat_features:
    temp=data.groupby(feature).count()/len(data)
    temp_df=temp[temp>0.01].index
    data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rare_var')
```

# Feature Scaling

```python
for feature in cat_features:
    labels_ordered=data.groupby([feature])['LotArea'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
feature_scale=[feature for feature in data.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[feature_scale])
```
